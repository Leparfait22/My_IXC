```python
import json
import pandas as pd

with open('locals.json', 'r', encoding='utf-8') as file_locals:
    locals_data = json.load(file_locals)
    
with open('devices.json', 'r', encoding='utf-8') as file_device:
    device_data = json.load(file_device)
    
for local in locals_data:
    if(local['name'] == 'Casa de Marcelo e Jessica '):
        casa_de_marcelo_jessica = local
        
print(casa_de_marcelo_jessica)
```

```python
devices_casa_marcelo_jessica = []

for device in device_data:
    device_local = device['local']
    if device_local['$oid'] == casa_de_marcelo_jessica['_id']['$oid']:
        devices_casa_marcelo_jessica.append(device)

devices_count = len(devices_casa_marcelo_jessica[0]['end_devices'])
print(f"Numero de devices : {devices_count}")
print()

for device in devices_casa_marcelo_jessica:
    print(device)
```

```python
casa_de_marcelo_jessica_copy = casa_de_marcelo_jessica.copy()
luciano_rooms = casa_de_marcelo_jessica_copy['rooms']

for room in luciano_rooms:
    # print(room)
    card = room['cards']
    print(card)
    if not card:
        print("card ausente")
        del room['cards']
    # print(entry)
    print('--------------------------------')
```
